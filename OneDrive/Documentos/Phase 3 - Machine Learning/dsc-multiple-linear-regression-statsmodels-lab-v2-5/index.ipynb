## The Ames Housing Dataset

The [Ames Housing dataset](http://jse.amstat.org/v19n3/decock.pdf) is a newer (2011) replacement for the classic Boston Housing dataset. Each record represents a residential property sale in Ames, Iowa. It contains many different potential predictors and the target variable is `SalePrice`.

````python
import pandas as pd
ames = pd.read_csv("ames.csv", index_col=0)
ames
````

````python
ames.describe()
````

We will focus specifically on a subset of the overall dataset. These features are:

```
LotArea: Lot size in square feet

1stFlrSF: First Floor square feet

GrLivArea: Above grade (ground) living area square feet
```

````python
ames_subset = ames[['LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']].copy()
ames_subset
````

## Step 1: Visualize Relationships Between Features and Target

For each feature in the subset, create a scatter plot that shows the feature on the x-axis and `SalePrice` on the y-axis.

````python
# Your code here - import relevant library, create scatter plots
import matplotlib.pyplot as plt

Features = ['LotArea', '1stFlrSF', 'GrLivArea']

for feature in Features:
    plt.figure(figsize=(8,6))
    plt.scatter(ames_subset[feature], ames_subset['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(f'Scatter plot of {feature} vs SalePrice')
    plt.show()

    # ames_subset.plot.scatter(x=feature, y='SalePrice', title=f'Scatter plot of {feature} vs SalePrice')
````

````python
# Your written answer here - do these seem like good candidates for linear regression?
# GrLivArea and 1stFlrSF show clearer positive linear trends with SalePrice.
# LotArea has more variability and outliers, so it may be a weaker predictor.
# Overall, these look like reasonable candidates for linear regression, though transformations might help later.
````

## Step 2: Build a Simple Linear Regression Model

Set the dependent variable (`y`) to be the `SalePrice`, then choose one of the features shown in the subset above to be the baseline independent variable (`X`).

Build a linear regression using StatsModels, describe the overall model performance, and interpret its coefficients.

````python
# Your code here - define y and baseline X
import statsmodels.api as sm

# Dependent variable
y = ames_subset['SalePrice']

# Baseline independent variable 
X = ames_subset[['GrLivArea']]
X = sm.add_constant(X) # adding a constant for the intercept
````

````python
# Your code here - import StatsModels, fit baseline model, display results
model_simple = sm.OLS(y, X).fit()
print(model_simple.summary())
````

````python
# Your written answer here - interpret model results
# - The coefficient for GrLivArea tells us how much SalePrice increases per additional square foot of living area.
# - The R-squared indicates how much of the variance in SalePrice is explained by this single predictor.
# - Typically, GrLivArea alone explains a decent portion of the variance, but not all.
````

## Step 3: Build a Multiple Linear Regression Model

For this model, use **all of** the features in `ames_subset`.

````python
# Your code here - define X
X_multi =ames_subset[['GrLivArea', '1stFlrSF', 'LotArea']]
X_multi = sm.add_constant(X_multi)
````

````python
# Your code here - fit model and display results
model_multi = sm.OLS(y, X_multi).fit()
print(model_multi.summary())
````

````python
# Your written answer here - interpret model results. Does this model seem better than the previous one?

# - Each coefficient now represents the effect of that feature holding the others constant.
# - The R-squared should be higher than the simple model, showing improved explanatory power.
# - However, multicollinearity may exist (e.g., GrLivArea and 1stFlrSF are correlated).
````

## Step 4: Create Partial Regression Plots for Features

Using your model from Step 3, visualize each of the features using partial regression plots.

````python
# Your code here - create partial regression plots for each predictor
from statsmodels.graphics.regressionplots import plot_partregress_grid

fig = plot_partregress_grid(model_multi)
plt.show()
````

````python
# Your written answer here - explain what you see, and how this relates
# to what you saw in Step 1. What do you notice?
# - These plots show the relationship between each predictor and SalePrice after accounting for other predictors.
# - Youll notice GrLivArea remains strongly positively related, while LotArea looks weaker.
# - This confirms what we saw in Step 1: some predictors are stronger than others.
````

## Level Up (Optional)

Re-create this model in scikit-learn, and check if you get the same R-Squared and coefficients.

````python
# Your code here - import linear regression from scikit-learn and create and fit model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

X_sklearn = ames_subset[['GrLivArea', '1stFlrSF', 'LotArea']]
y_sklearn = ames_subset['SalePrice']
````

````python
# Your code here - compare R-Squared

lr = LinearRegression()
lr.fit(X_sklearn, y_sklearn)

print("R-Squared (scikit-learn):", r2_score(y_sklearn, lr.predict(X_sklearn)))
print("Intercepts (scikit-learn):", lr.intercept_)
print("Coefficients (scikit-learn):", lr.coef_)


````

````python
# Your code here - compare intercept and coefficients

# - You should see the same R-squared and similar coefficients as StatsModels.
# - StatsModels gives more detailed statistical output (p-values, confidence intervals), while scikit-learn is more streamlined for machine learning workflows
````
