```python
"""
1. Save full names( first & last)
2. uppercase/capitalise the name
3. send a message ' Welcome-to-Moringa'
"""
fullname = None

def get_name():
  firstname = str(input('Firstname'))
  lastname = str(input('Lastname'))
  fullname = firstname + ' ' + lastname
  return fullname

def capitalize():
  return fullname.upper()

```

```python
class Student:
    def __init__(self, first_name, last_name):
        # attributes
        self.first__name = first_name
        self.last_name = last_name

    def capitalize(self):
        return f"{self.first__name.upper()} {self.last_name.upper()}"  

    def welcome(self):
        return f'Dear {self.first__name}, Welcome to Moringa!' 

```

```python
student1 = Student('Geoffrey', 'Kemboi')   # first instance of the student class
student1.welcome()
```

```python
type(student1)
```

```python
student1 = Student('geoffrey', 'kemboi')  # second instance of the student class
student2 = Student('angela', 'mukami')    # third instance of the student class
student3 = Student('Peter', 'Kiarie')     # fourth instance of the student class
student4 = Student('faith', 'koech')      # fifth instance of the student class
student5 = Student('duncan', 'nyaga')     # sixth instance of the student class
student6 = Student('mercy', 'kangangi')   # seventh instance of the student class
```
