````python
class Student:

  def __init__(self, name, age, grade):
    # initializing attributes of an instance 
    self.name = name
    self.age = age
    self.grade = grade    # range from 0 - 100

  def get_grade(self):
    return self.grade
  
class Course:
  def __init__(self, name, max_students):
    self.name = name
    self.max_students = max_students
    self.students = []

  def add_student(self, student):
    if len(self.students) < self.max_students:
      self.students.append(student)
      return True
    return False
  
  def get_average_grade(self):
    value = 0
    for student in self.students:
      value += student.get_grade()

    return value/len(self.students)
  
# instantiating
s1 = Student("Tim",19, 95)
s2 = Student("Dom",17, 75)
s3 = Student("Geoff", 19, 87)

course = Course("Science", 2)
course.add_student(s1)
course.add_student(s2)
print(course.add_student(s3))
print(course.get_average_grade()) 


````

````python

## ðŸ§© What They Are
# - In Python, when you see `def` inside a class, youâ€™re defining a **method** (a function that belongs to a class).
# - Methods describe **behaviors** of the objects (instances) created from that class.
# - They always take `self` as the first parameter, which refers to the specific instance calling the method.

# ---

## ðŸ“š Example Context
# Imagine you have a `Course` class:

# ```python
class Course:
    def __init__(self, name, max_students):
        self.name = name
        self.max_students = max_students
        self.students = []

    def add_student(self, student):
        if len(self.students) < self.max_students:
            self.students.append(student)
            return True
        return False

    def get_average_grade(self):
        value = 0
        for student in self.students:
            value += student.get_grade()
        return value / len(self.students)
# ```

# ---

## ðŸ” Breakdown of Each Method

### 1. `add_student(self, student)`
# - **Purpose**: Adds a student to the course if thereâ€™s room.
# - **Logic**:
#   - Checks if the current number of students (`len(self.students)`) is less than the maximum allowed (`self.max_students`).
#   - If yes â†’ appends the new student to the list and returns `True`.
#   - If no â†’ returns `False` (meaning the course is full).
# - **Behavior**: Controls enrollment.

### 2. `get_average_grade(self)`
# - **Purpose**: Calculates the average grade of all students in the course.
# - **Logic**:
#   - Starts with `value = 0`.
#   - Loops through each student in `self.students`.
#   - Calls `student.get_grade()` (which must be a method of the `Student` class).
#   - Adds up all grades, then divides by the number of students.
# - **Behavior**: Computes statistics about the course.

# ---

## âœ¨ Analogy
# Think of the class as a **school course**:
# - `add_student()` = The **enrollment desk** ðŸ“ (decides if a student can join).
# - `get_average_grade()` = The **report generator** ðŸ“Š (calculates how the class is performing).

# ---

## ðŸ Demo in Action
```python
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def get_grade(self):
        return self.grade

course = Course("Math", 2)

s1 = Student("Tim", 95)
s2 = Student("Dom", 75)
s3 = Student("Geoff", 87)

print(course.add_student(s1))  # True
print(course.add_student(s2))  # True
print(course.add_student(s3))  # False (course full)

print(course.get_average_grade())  # (95 + 75) / 2 = 85.0
```

---

âœ… So: those are **methods inside a class** that define what the object can *do*. They let your objects go beyond just storing data â€” they can perform actions and calculations.  
.
````
